````python
from sklearn.feature_extraction.text import CountVectorizer
count_vector = CountVectorizer(stop_words="english",token_pattern=r'\b[^\d\W]+\b') #set the variable

count_vector.fit(df.sms_message) #fit the function
count_vector.get_feature_names() #get the outputs
````

````python
doc_array = count_vector.transform(df.sms_message)
````
